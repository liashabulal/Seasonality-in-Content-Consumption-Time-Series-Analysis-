```python
import pandas as pd
```

```python
import numpy as np
```

```python
dates = pd.date_range('2020-01-01' ,'2025-01-01', freq = 'MS') 
```

```python
len(dates)
```

```python
categories = ['News', 'Fashion', 'Tech', 'Travel']
print(categories)
```

```python
data = []
```

```python
for cat in categories :
    base = np.random.randint(10000,30000)
    seasonality= np.sin(np.linspace(0, 12*np.pi, len(dates)))*3000
    noise = np.random.normal(0 , 1500 , len(dates))
    views = base + seasonality + noise

    for d, v in zip(dates, views):
        data.append([d , cat , max(0,int(v))])
     
```

```python
#np.linspace will create random values , and the sine function will create values between -1 to +1 and this will be multipled with 3000 to visibily plot it
```

```python
# simulates real world randomness 
```

```python
syn_data= pd.DataFrame(data, columns=['Date', 'Category', 'Views'])
```
